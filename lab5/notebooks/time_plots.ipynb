```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Redshift CSV

```python
redshift = pd.DataFrame(
    {
        'SF1': 
            { 
                'Q1': [10.17, 0.312, 0.31],
                'Q5': [26.358, 0.314, 0.325] 
            },
        'SF10': 
            { 
                'Q1': [29.22, 0.2, 0.2],
                'Q5': [34.38, 0.22, 0.22] 
            },
        'SF100':
            {
                'Q1': [21.69, 0.22, 0.22],
                'Q5': [48.01, 0.21, 0.21] 
            }
    }
)
```

```python
xlab = ['Run 1', 'Run 2', 'Run3']
plt.plot(xlab, redshift['SF1']['Q1'], '-ob', label='SF1 - Query 1')
plt.plot(xlab, redshift['SF1']['Q5'], '-vb', label='SF1 - Query 5')

plt.plot(xlab, redshift['SF10']['Q1'], '-or', label='SF10 - Query 1')
plt.plot(xlab, redshift['SF10']['Q5'], '-vr', label='SF10 - Query 5')

plt.plot(xlab, redshift['SF100']['Q1'], '-og', label='SF100 - Query 1')
plt.plot(xlab, redshift['SF100']['Q5'], '-vg', label='SF100 - Query 5')

plt.legend(loc='best')
plt.title('Redshift Performance for .csv files')
plt.show()
```
